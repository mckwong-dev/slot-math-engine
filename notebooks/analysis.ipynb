## Tuning comparison example

```python
import copy

variant = copy.deepcopy(config)
variant['paytable']['line']['A'][5] = 120  # example tuning change

base_report = run_simulation(config, spins=100_000, seed=777)
variant_report = run_simulation(variant, spins=100_000, seed=777)

{
    'base_rtp': base_report['rtp'],
    'variant_rtp': variant_report['rtp'],
    'base_hit_frequency': base_report['hit_frequency'],
    'variant_hit_frequency': variant_report['hit_frequency'],
    'base_sd': base_report['standard_deviation'],
    'variant_sd': variant_report['standard_deviation'],
}
```
